```python
import sys, time
import pandas as pd
import datetime as dt
from IPython.display import display
```

```python
FILE_NAME = "output.gml"
```

```python
gender_symbol = []
```

```python
for x, y in enumerate(df['label']):
#     str = df['label'][x] + '(' + df[''] 
#     df['label'] = df['label'] += df['gen']
    string_s = df['label'][x] + ' (' + str(df['gender description'][x]) + ')'
    
    if df['gender description'][x] == 'male':
        gender_symbol.append('triangle')
    else:
        gender_symbol.append('oval')
        
    second_str = 'Chap. ' + str(df['chapter_left'][x]) + ' - ' + df['action description'][x]
        
    df['action description'].loc[[x]] = second_str
    
    df['label'][x] = string_s
    
```

```python
df = df.assign(genderSymbol=gender_symbol)
```

```python
df.head()
```

```python
#Utility functions
def progress(v):
    v = str(v)
    sys.stdout.flush()
    sys.stdout.write('\r')
    sys.stdout.flush()
    sys.stdout.write(v)
```

```python
f = open(FILE_NAME, "w")
#helpers
s = " "
ss = s+s
sss = s+s+s
ssss = s+s+s+s
nl = "\n"

#loop helpers
added = []
ind = 0

#Root node
f.write("graph"+nl)
f.write("["+nl)



#Write a node
def write_node(r):
    f.write( ss + "node" + nl)
    f.write( ss + "[" + nl)
    f.write( ssss + "id" + s + str(r['source']) + nl)
    f.write( ssss + "label" + s + '"' + str(r['label']) + '"' + nl)
    f.write( ssss + "graphics" + nl + ssss + "[" + s + "type" + s  + '"' + str(r['genderSymbol']) + '"'\
            +nl + ssss +\
            ']' 
            + nl)
    f.write( ssss + "LabelGraphics" + nl + ssss + "[" + s + "anchor" + s  + '"' + "s" + '"'\
            + nl  + ssss  + ']'\
            + nl)
    f.write( ss + "]"+ nl)


#Write an edge
def write_edge(r):
    f.write( ss + "edge" + nl)
    f.write( ss + "[" + nl)
    f.write( ssss + "source" + s + str(r['source'])  + nl)
    f.write( ssss + "target" + s  + str(r['target'])  + nl)
#     f.write( ssss + "graphics" + nl + ssss + "[" + s + "style" + s  + '"' + "dashed" + '"'\
#             +nl + ssss +\
#             ']' 
#             + nl)
    f.write( ssss + "LabelGraphics" + nl + ssss + "[" + s + "text" + s + '"' + str(r['action description']) \
            + '"' +\
            nl + ssss + ']'  
            + nl)
    f.write( ss + "]"+ nl)


#Generate nodes
for i, r in df.iterrows():
    #increment, as index not reliable
    ind += 1
    #Check for duplicates
    if (r['source'] not in added):
        #Add to list
        added.append(r['source'])
        write_node(r)
    #print the progress    
    progress(ind)

print(nl+"Printing nodes over")

#flush index
ind = 0    
#Generate edges    
for i, r in df.iterrows():
    #increment, as index not reliable
    ind += 1
#     if(r['source'] < r['target']):
    write_edge(r)
    #print the progress            
    progress(ind)

print(nl+"Printing nodes and edges over")

#closing node
f.write("]"+nl)
f.close()
```
